## Room Impulse Responses Captured by Circular Microphone Array

In the following room impulse responses which have been captured sequentially on a circle with $R=1$ m in Auditorium 3 of Telekom Innovation Laboratories, Technical University of Berlin are used as an example of an spatio-temporal impulse response. The picture shows the measurement setup.

![Measurement of room impulse responses](measurement_auditorium3.jpg)

The loudspeaker emitted a sweep signal which was captured by the microphone mounted at the end of an arm. After each measurement the arm was rotated by one degree resulting in a total of 360 room impulse responses.

### Load Dataset

First the impulse responses are loaded

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

data = np.load('auditorium_3_aequatorial_IR_truncated.npz')

fs = data['fs']
phi = data['phi']
ir = data['ir']
```

### Plot of Spatio-Temporal Impulse Response

The magnitude of the spatio-temporal impulse responses $h(\theta, t)$ is plotted on a logarithmic scale (dB) for ease of illustration.

```python
t = (1/fs) * np.arange(ir.shape[1])

plt.figure(figsize=(10,10))
plt.pcolormesh(phi, t, 20*np.log10(np.abs(ir.T)))
plt.xlabel(r'$\theta$ in deg')
plt.ylabel('$t$ in s')
plt.title(r'$|h(\theta, t)|$ in dB')
plt.clim([-50, 0])
plt.colorbar();
```

One particular impulse response for one particular angle $\theta$ is plotted in the following.

```python
idx = 210

plt.figure(figsize=(10, 5))
plt.plot(t, ir[idx,:])
plt.title(r'$\theta = %d^\mathrm{o}$' %phi[idx])
plt.xlabel('$t$ in s')
plt.ylabel(r'$h(\theta, t)$')
plt.grid()
```
