## Example - Moving Scatterer

As example for a time-variant acoustic scenario, the influence of a scatter moving through the line-of-sight between a loudspeaker and a microphone is considered. The respective time-varying room impulse response was measured while the experimenter walked through the line-of-sight between the source (loudspeaker) and the receiver (microphone) both placed in room W8202 of the Institute of Communications Engineering. The direct sound and the reflection from the floor were obstructed by the body, causing a low-pass filter-like effect.

### Time-Varying Impulse Response

The time-varying impulse response $h(t, t-\tau)$ is loaded and plotted. The time $t$ denotes the absolute time the impulse response was captured at.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio


data = np.load('timevariant_IR_scatterer.npz')
fs = data['fs']
ir = data['ir']
hop = data['hop']

tau = 1/fs * np.arange(ir.shape[0])
t = (hop/fs) * np.arange(ir.shape[1])

plt.figure(figsize=(12, 8))
plt.pcolormesh(t, tau, ir, shading='auto')
plt.xlabel('$t$ in s')
plt.ylabel(r'$\tau$ in s')
plt.axis([t[0], t[-1], 0, 0.05])
plt.clim([-2, 2])
plt.colorbar()
```

Now one particular impulse response captured at one particular time-instant is plotted for illustration

```python
idx = 100

plt.figure(figsize=(10, 10))
plt.plot(tau, ir[:, idx])
plt.title('Impulse response at $t=$ %d s' % (hop/fs*idx))
plt.xlabel(r'$\tau$ in s')
plt.ylabel(r'$h(t, t-\tau)$')
plt.grid()
```
