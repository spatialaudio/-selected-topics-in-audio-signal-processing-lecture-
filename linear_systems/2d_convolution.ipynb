## Image Filtering

As an example of multidimensional convolution, the filtering of an image is illustrated in the following. First a function is defined to convert an color image to gray scale suitable for further processing.

```python
%matplotlib inline
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

An example image from `scipy` is loaded, converted to grayscale and plotted.

```python
q = rgb2gray(misc.face())
plt.imshow(q, cmap = 'gray');
```

The image is convolved with a two-dimensional rectangular function. The resulting image is plotted.

```python
n = 50
window = 1/n**2 * np.ones((n,n))
p = convolve2d(q, window, mode="same", boundary="symm")
        
plt.imshow(p, cmap = 'gray');
```

The Fourier transform of the rectangular window is sinc-shaped. The sinc function has a low-pass character which results in an attenuation of the high-frequencies in the image. The high-frequencies represent the structures with fine detail in the image. Overall, this explains the blurry result.
